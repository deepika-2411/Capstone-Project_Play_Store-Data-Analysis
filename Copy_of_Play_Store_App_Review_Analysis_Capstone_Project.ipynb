```python
playapp_df_1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/excel/ User Reviews.csv')
#playapp_df_1.set_index('App',inplace=True)
playapp_df_1
```

```python
playapp_df_1.dropna(axis=0, how='all', subset=['Translated_Review'], inplace=True)
playapp_df_1
```

```python
new_df=playapp_df_1.iloc[:,1:2]
new_df
```

```python
count=0
for item in new_df['Translated_Review']:
  count=count+len(item.split())
print("Total number of words in Translated_Review=",count)
print("Total number of words in Translated_Review=",round(count/37427),'words.')
```

```python
playapp_df_1.dropna(axis=0, how='all', subset=['Sentiment'], inplace=True)
playapp_df_1
```

```python
count=0
for word in playapp_df_1['Sentiment']:
  if word=='Positive':
    count=count+1
print("Total number of positive sentiment in sentiment=",count)
```

```python
count_sentiment = playapp_df_1.groupby(["Sentiment"])["Sentiment"].count()
count_sentiment
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
plt.plot(count_sentiment)
plt.title('Different user sentiment')
plt.ylabel('total no. of different user sentiment')
plt.xlabel('Sentiment')
```

```python
count_sentiment.plot(kind='bar')
plt.title('Different user sentiment')
plt.ylabel('total no. of different user sentiment')
plt.xlabel('Sentiment')
```

```python
merged_RJ_df = pd.merge(playapp_df, playapp_df_1, how='left',left_on=['App'],right_on='App')
merged_RJ_df
```

```python
merged_RJ_df.dropna(axis=0, how='all', subset=['Sentiment_Polarity'], inplace=True)
merged_RJ_df
```

```python
a=merged_RJ_df[ (merged_RJ_df['Sentiment_Polarity'] > 0.5) & (merged_RJ_df['Sentiment_Subjectivity'] > 0.5)].reset_index(drop=True)
a
```

```python
b=a.drop(['year','Rating','Reviews','Size','Installs', 'Type', 'Price', 'Content Rating', 'Genres', 'Last Updated', 'Current Ver', 'Android Ver','Translated_Review','Sentiment'],axis=1)
c=b.head(20)
```

```python
c.plot(x='Category', y='Sentiment_Polarity',figsize=(10,5),kind="bar")
```
